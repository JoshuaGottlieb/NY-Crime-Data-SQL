## Load Libraries and Data

```python
import numpy as np
import pandas as pd
```

```python
population_data_csv = '../data/raw/New_York_City_Population_by_Borough__1950_-_2040_20241014.csv'
```

```python
df = pd.read_csv(population_data_csv)
df
```

```python
# Drop first column and share of NYC total columns
df.drop(['Age Group'] + [x for x in df.columns if 'share' in x], axis = 1, inplace = True)
```
